```
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

H.keys()
plt.figure(figsize=(26,26))
LH = []
kh = H.keys()
LH.append(kh);            
for i,kle in enumerate(H.keys()) :
  hloss = H[kle]
  plt.plot(np.log(hloss), alpha=0.5)
  hloss = np.reshape(hloss,(1,len(hloss)))
  LH.append(hloss[0])               
plt.legend(H.keys(), framealpha=0.3, loc=0)
plt.xlabel('epoch')    
plt.ylabel('log(loss)')
#plt.savefig("/content/drive/MyDrive/TRIED/Projet Long/loss.png")#,dpi=900)
```
